###2. Mejorando Mandelbrot con pyCuda
Nuevamente, trataremos de mejorar el tiempo en obtener el conjunto de Mandelbrot, ahora usndo pyCuDA.
Comenzamos por importar las librerias:


```python
import numpy as np
import time
from pylab import imshow, show
from timeit import default_timer as timer

import pycuda.driver as drv
import pycuda.tools
import pycuda.autoinit
from pycuda.compiler import SourceModule
import pycuda.gpuarray as gpuarray
```

Definimos una función que determina si un número complejo pertenece al conjunto de Mandelbrot: si después de un número de iteraciones, `max_iters`, la magnitud del número complejo se sale de la frontera, pertenece al conjunto.

```python
def mandel(c, z, max_iters):
    """Calculate z using numpy on the GPU via gpuarray"""
    outputg = gpuarray.to_gpu(np.resize(np.array(0,), c.shape).astype(np.int32))
    zg = gpuarray.to_gpu(z.astype(np.complex64))
    cg = gpuarray.to_gpu(c.astype(np.complex64))
    # 2.0 as an array
    twosg = gpuarray.to_gpu(np.array([2.0]*zg.size).astype(np.float32))
    # 0+0j as an array
    cmplx0sg = gpuarray.to_gpu(np.array([0+0j]*zg.size).astype(np.complex64))
    # for abs_zg > twosg result
    comparison_result = gpuarray.to_gpu(np.array([False]*zg.size).astype(np.bool))
    # we'll add 1 to iterg after each iteration
    iterg = gpuarray.to_gpu(np.array([0]*zg.size).astype(np.int32))

    for iter in range(max_iters):
        zg = zg*zg + cg

        # abs returns a complex (rather than a float) from the complex
        # input where the real component is the absolute value (which
        # looks like a bug) so I take the .real after abs()
        abs_zg = abs(zg).real

        comparison_result = abs_zg > twosg
        cg = gpuarray.if_positive(comparison_result, cmplx0sg, cg)
        zg = gpuarray.if_positive(comparison_result, cmplx0sg, zg)
        outputg = gpuarray.if_positive(comparison_result, iterg, outputg)
        iterg = iterg + 1
    output = outputg.get()
    return output
```

La siguiente función genera el fractal a partir de los pixeles de una imagen:

```python
def create_fractal(min_x, max_x, min_y, max_y, image, iters):
    height = image.shape[0]
    width = image.shape[1]
    
    pixel_size_x = (max_x - min_x) / width
    pixel_size_y = (max_y - min_y) / height
    
    xx = np.arange(min_x, max_x, pixel_size_x)
    yy = np.arange(min_y, max_y, pixel_size_y) *1j
    yy = yy.astype(np.complex64)
    
    c = np.ravel(xx+yy[:, np.newaxis]).astype(np.complex64)
    z = np.zeros(c.shape, np.complex64)
    
#     for x in range(width):
#         real = min_x + x * pixel_size_x
#         for y in range(height):
#             imag = min_y + y * pixel_size_y
#     return c, z
    color = mandel(c, z, iters)
    return color.reshape(image.shape)
#     image = color.reshape(image.shape)
```

Creamos un arreglo que contendrá la imagen y medimos el tiempo que tarda en generarse el conjunto de Mandelbrot después de 20 iteraciones:

```python
image = np.zeros((1024,1536), dtype=np.uint8)
start = timer()
image = create_fractal(-2.0, 1.0, -1.0, 1.0, image, 20)
dt = timer() - start

print "Mandelbrot creado en %f s" % dt

imshow(image, extent=(-2,1,-1,1))
show()
```

```python
image = np.zeros((1024,1536), dtype=np.uint8)
start = timer()
image = create_fractal(-1.82, -1.72, -0.06, 0.06, image, 20)
dt = timer() - start

print "Mandelbrot creado en %f s" % dt

imshow(image, extent=(-1.82, -1.72, -0.06, 0.06))
show()
```

Comparado con el tiempo de ejecución de los codigos de las libretas pasadas, donde habiamos encontrado un aumento en la eficiencia de aproximadamente 3.5 veces a usar las instrucciones vectorizadas de `Numpy`, con respecto a una implementación serial usando Python, vemos que el tiempo se reduce todavia más si usamos el GPU obteniendo del orden de un incremento de 2.4 respecto a la implementacion vectorizada y ¡hasta 8 veces mejor tiempo de ejecucion que en el caso serial!
