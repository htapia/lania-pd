###2. Numba autojit y cuda
Ahora presentaremos otra herramienta que permite accesar al GPU desde Python. A diferencia de pyCUDA, esta implementación crea codigo compilado al momento `just in time` o `jit` a partir de codigo de Python, abtrayendo aun mas el modelo de programacion CUDA sobre todo en la escritura de la función _kernel_.

Utilizaremos primero el codigo usado en la libreta pasada donde empelamos puro Python par ejemplificar el uso de `numba`.

```python
import numpy as np
import time
from pylab import imshow, show
from timeit import default_timer as timer

from numba import autojit
```

Para usar estas funciones LLVM  unicamente antecedemos `@autojit` a estas definiciones:

```python
@autojit
def mandel(x, y, max_iters):
    c = complex(x,y)
    z = 0.0j
    for i in range(max_iters):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return max_iters
```

```python
@autojit
def create_fractal(min_x, max_x, min_y, max_y, image, iters):
  height = image.shape[0]
  width = image.shape[1]

  pixel_size_x = (max_x - min_x) / width
  pixel_size_y = (max_y - min_y) / height
    
  for x in range(width):
    real = min_x + x * pixel_size_x
    for y in range(height):
      imag = min_y + y * pixel_size_y
      color = mandel(real, imag, iters)
      image[y, x] = color
```

El resto es identico a los codigos anteriores:

```python
image = np.zeros((1024,1536), dtype=np.uint8)
start = timer()
create_fractal(-2.0, 1.0, -1.0, 1.0, image, 20)
dt = timer() - start

print "Mandelbrot creado en %f s" % dt

imshow(image, extent=(-2,1,-1,1), aspect='auto')
show()
```

```python
image = np.zeros((1024,1536), dtype=np.uint8)
start = timer()
create_fractal(-1.82, -1.80, -0.001, 0.001, image, 20)
dt = timer() - start

print "Mandelbrot creado en %f s" % dt

imshow(image, extent=(-1.82, -1.72, -0.06, 0.06))
show()
```

```python
%timeit create_fractal(-1.82, -1.80, -0.001, 0.001, image, 20)
```

¡Usando autojit de numba mejoramos el tiempo de ejecucion en unas 12 veces comparadas con el uso de pyCUDA! Esto es debido a que autojit crea codigo compilado que es mucho mas rápido que el codigo interpretado de que incluso se ejecuta con pyCUDA.

Ahora usaremos otra librearia que si compila sobre el GPU.

```python
from numbapro import cuda
from numba import *
```

```python
def mandel(x, y, max_iters):
  """
    Given the real and imaginary parts of a complex number,
    determine if it is a candidate for membership in the Mandelbrot
    set given a fixed number of iterations.
  """
  c = complex(x, y)
  z = 0.0j
  for i in range(max_iters):
    z = z*z + c
    if (z.real*z.real + z.imag*z.imag) >= 4:
      return i

  return max_iters
```

```python
mandel_gpu = cuda.jit(restype = uint32, \
                      argtypes = [f8, f8, uint32], \
                      device = True)(mandel)                      
```

```python
mandel_gpu
```

```python
@cuda.jit(argtypes=[f8, f8, f8, f8, uint8[:,:], uint32])
def mandel_kernel(min_x, max_x, min_y, max_y, image, iters):
  height = image.shape[0]
  width = image.shape[1]

  pixel_size_x = (max_x - min_x) / width
  pixel_size_y = (max_y - min_y) / height
    
  startX, startY = cuda.grid(2)
  gridX = cuda.gridDim.x * cuda.blockDim.x;
  gridY = cuda.gridDim.y * cuda.blockDim.y;
    
  for x in range(startX, width, gridX):
    real = min_x + x * pixel_size_x
    for y in range(startY, height, gridY):
      imag = min_y + y * pixel_size_y
      image[y, x] = mandel_gpu(real, imag, iters)
```

```python
gimage = np.zeros((1024,1536), dtype=np.uint8)
blockDim = (32, 8)
gridDim = (32,16)

start = timer()
d_image = cuda.to_device(gimage)
mandel_kernel[gridDim, blockDim](-2.0, 1.0, -1.0, 1.0, d_image, 20)
d_image.to_host()
dt = timer() - start

print "Mandelbrot creado en %f s" % dt

imshow(gimage, extent=(-2.0, 1.0, -1.0, 1.0))
show()
```

Resultado, usando numbapro se reduce aun mas el tiempo de ejecución.
