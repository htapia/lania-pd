###2. Generando Mandelbrot con Numpy.
En esta libreta utilizaremos Python y pyCUDA para generar imagenes hoy populares, conocidas como fractales. En particular implementaremos la generación del conjunto de Mandelbrot y el conjunto de Julia, utilizando Python comun (con NumPy) y pyCUDA.

Comenzaremos construyendo el conjunto de Mandelbrot usando Python:

```python
import numpy as np
import time
from pylab import imshow, show
from timeit import default_timer as timer
```

Definimos una función que determina si un número complejo pertenece al conjunto de Mandelbrot: si después de un número de iteraciones, `max_iters`, la magnitud del número complejo se sale de la frontera, pertenece al conjunto.

```python
def mandel(x, y, max_iters):
    c = complex(x,y)
    z = 0.0j
    for i in range(max_iters):
        z = z*z + c
        if (z.real*z.real + z.imag*z.imag) >= 4:
            return i
    return max_iters
```

La siguiente función genera el fractal a partir de los pixeles de una imagen:

```python
def create_fractal(min_x, max_x, min_y, max_y, image, iters):
    height = image.shape[0]
    width = image.shape[1]
    
    pixel_size_x = (max_x - min_x) / width
    pixel_size_y = (max_y - min_y) / height
    
    for x in range(width):
        real = min_x + x * pixel_size_x
        for y in range(height):
            imag = min_y + y * pixel_size_y
            color = mandel(real, imag, iters)
            image[y, x] = color            
```

Creamos un arreglo que contendrá la imagen y medimos el tiempo que tarda en generarse el conjunto de Mandelbrot después de 20 iteraciones:

```python
image = np.zeros((1024,1536), dtype=np.uint8)
start = timer()
create_fractal(-2.0, 1.0, -1.0, 1.0, image, 20)
dt = timer() - start

print "Mandelbrot creado en %f s" % dt
```

Ahora podemos visualizar el resultado:

```python
imshow(image)
```

```python
imshow(image, extent=(-2,1,-1,1))
show()
```

```python
image = np.zeros((1024,1536), dtype=np.uint8)
start = timer()
create_fractal(-1.82, -1.72, -0.06, 0.06, image, 20)
dt = timer() - start

print "Mandelbrot creado en %f s" % dt
```
