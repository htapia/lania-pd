###1. pyCUDA y analisis de imagenes
En esta libreta aplicaremos las herramientas que hemos desarrollado para una aplicación muy concreta: desarrollaremos un código que, dado un archivo con datos reconstruye la imagen original. Este es un ejemplo muy simple de un algoritmo del tipo Reverse Edge Detection.

Como siempre, primero llamaremos las librerías necesarias para invocar pyCUDA:

```python
import pycuda.driver as cuda
import pycuda.autoinit
from pycuda.compiler import SourceModule
```

A continuación mostramos los datos sin reconstruir:

```python
edge = np.fromfile('codes/learnCUDA/input_files/edge256x192.dat',dtype=int32,sep=' ')
HEIGHT,WIDTH = edge[0:2] # dimensions of image
edge = edge[2:]
# eh=edge[::-1].reshape(WIDTH,HEIGHT)
eh=edge.reshape(WIDTH,HEIGHT)
```

```python
# initialize input array to zeroes in the host
h_input = np.zeros([WIDTH+2,HEIGHT+2])
```

```python
# copy data to input leaving the edges clear in zeros to have a halo
# h_input[1:WIDTH+1,1:HEIGHT+1] = eh[:,::-1].astype(np.float32)
h_input[1:WIDTH+1,1:HEIGHT+1] = eh.astype(np.float32)
```

```python
# loop on gpu up to iterations
# run kernel
GRID_W = 16
GRID_H = 12
BLOCK_W = 16
BLOCK_H = 16
```

```python
mod = SourceModule("""
__global__ void inverseEdgeDetect1D_col(float *d_output, float *d_input, float *d_edge, int WIDTH)
{
    int col, row;
    int idx, idx_south, idx_north, idx_west, idx_east;
    int numcols = WIDTH + 2;
    
    // col = blockIdx.x*blockDim.x + threadIdx.x;
    row = blockIdx.x*blockDim.x + threadIdx.x + 1;
    
    for (col = 1; col <= WIDTH; col++) {
    
        idx = row * numcols + col;
        idx_south = (row - 1) * numcols + col;
        idx_north = (row + 1) * numcols + col;        
        
        idx_west = row * numcols + (col - 1);
        idx_east = row * numcols + (col + 1);
        
        d_output[idx] = (d_input[idx_south] + d_input[idx_west] + \
                         d_input[idx_north] + d_input[idx_east] - \
                         d_edge[idx])*0.25;
    }
}
""")

inverseEdgeDetect = mod.get_function("inverseEdgeDetect1D_col")
```

```python
# allocate memory on device
d_input = cuda.mem_alloc(h_input.nbytes)
d_output = cuda.mem_alloc(h_input.nbytes)
d_edge = cuda.mem_alloc(h_input.nbytes)

# copy input to all device arrays
cuda.memcpy_htod(d_input, h_input)
cuda.memcpy_htod(d_output, h_input)
cuda.memcpy_htod(d_edge, h_input)

gpu_output = numpy.empty_like(h_input)

ITER = 1

for i in range(ITER):
    
    inverseEdgeDetect(d_output, d_input, d_edge, WIDTH, 
                      grid = (GRID_H, 1, 1), block = (BLOCK_H, 1, 1))
#      probably need to sync threads
    cuda.Context.synchronize()
#     copy data back from the device
    cuda.memcpy_dtoh(gpu_output, d_output)
    # copy new data to the input on device
    cuda.memcpy_htod(d_input, gpu_output)

#     cuda.Context.synchronize()

subplot(121)
imshow(h_input,aspect='equal')
subplot(122)
imshow(gpu_output,aspect='equal');
```

```python
thresh = 255.0
xmax = np.abs(gpu_output).max()
xmin = np.abs(gpu_output).min()
```

```python
tmp = gpu_output
if (xmin < 0 or xmax > thresh):
    tmp = np.int32(thresh*np.abs((tmp-xmin)/(xmax-xmin))+0.5)
else:
    tmp = np.int32(np.abs(tmp) + 0.5)
grey = np.int32(thresh * np.sqrt(tmp/thresh))
```
