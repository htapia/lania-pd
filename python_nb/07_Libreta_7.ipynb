###1. Difusión en 2D

Debemos _mapear_ el problema al hardware. En el modelo de CUDA el indexado de los nodos los agrupa de manera natural y cada _hilo_ opera en un nodo de la malla.

```python
/*** Heat equation with FD in global memory ***/
#include <iostream>
#include <fstream>
#include <cmath>
#include <sys/time.h>
#include <stdio.h>
#include <string.h>

void checkErrors(char *label)
{
// we need to synchronise first to catch errors due to
// asynchroneous operations that would otherwise
// potentially go unnoticed
cudaError_t err;
err = cudaThreadSynchronize();
if (err != cudaSuccess)
{
char *e = (char*) cudaGetErrorString(err);
fprintf(stderr, "CUDA Error: %s (at %s)\n", e, label);
}
err = cudaGetLastError();
if (err != cudaSuccess)
{
char *e = (char*) cudaGetErrorString(err);
fprintf(stderr, "CUDA Error: %s (at %s)\n", e, label);
}
}

double get_time() 
{  struct timeval tim;
  cudaThreadSynchronize();
  gettimeofday(&tim, NULL);
  return (double) tim.tv_sec+(tim.tv_usec/1000000.0);
}

// GPU kernels
__global__ void copy_array (float *u, float *u_prev, int N, int BSZ)
{	/******* write your kernel here! ***/
	int x = blockIdx.x*blockDim.x + threadIdx.x;
	int y = blockIdx.y*blockDim.y + threadIdx.y;
	if ((x<N) && (y<N)) {
	// printf("%d\t%d\t%d\n", i, j , i + j * blockDim.x * gridDim.x);
		// int idx = i + j * blockDim.x * gridDim.x;
		// int idx = N*j + i;
		int idx = x + y * blockDim.x*gridDim.x;
		u_prev[idx] = u[idx];
	}
	// __syncthreads();
}

__global__ void update (float *u, float *u_prev, int N, float h, float dt, float alpha, int BSZ)
{	/***** write your kernel here! ***/	
	int x = blockIdx.x*blockDim.x + threadIdx.x;
	int y = blockIdx.y*blockDim.y + threadIdx.y;
	// if(x*(N-x-1)*y*(N-y-1)!=0) {
	if (x>0 && (x<N-1) && y>0 && (y<N-1)) {
		int idx = x + y * blockDim.x*gridDim.x;
		// if ( (x==0) || (y==0)) {u[idx] = 200.0f;}
		u[idx] = u_prev[idx] + alpha*dt/(h*h) * (u_prev[idx+1] + u_prev[idx-1] + u_prev[idx+N] + u_prev[idx-N] - 4*u_prev[idx]);
	}
}

int main()
{
	// Allocate in CPU
	int N = 512;
	int BLOCKSIZE = 16;

	float xmin 	= 0.0f;
	float xmax 	= 3.5f;
	float ymin 	= 0.0f;
	//float ymax 	= 2.0f;
	float h   	= (xmax-xmin)/(N-1);
	float dt	= 0.00001f;	
	float alpha	= 0.645f;
	float time 	= 0.4f;

	int steps = ceil(time/dt);
	int I;

	float *x  	= new float[N*N]; 
	float *y  	= new float[N*N]; 
	float *u  	= new float[N*N];
	// float *u_h 	= new float[N*N];
	// float *u_prev  	= new float[N*N];


    std::cout<<"N = "<<N<<std::endl;

	// Generate mesh and intial condition
	for (int j=0; j<N; j++)
	{	for (int i=0; i<N; i++)
		{	I = N*j + i;
			x[I] = xmin + h*i;
			y[I] = ymin + h*j;
			u[I] = 0.0f;
			if ( (i==0) || (j==0)) 
				{u[I] = 200.0f;}
		}
	}

	// Save initial condition
	std::ofstream temperature_ic("temperature_ic_global.txt");
	for (int j=0; j<N; j++)
	{	for (int i=0; i<N; i++)
		{	I = N*j + i;
			temperature_ic<<x[I]<<"\t"<<y[I]<<"\t"<<u[I]<<std::endl;
		}
		temperature_ic<<"\n";
	}

	temperature_ic.close();

	// Allocate in GPU
	// float *u, float *u_prev, int N, float h, float dt, float alpha, int BSZ

	double start = get_time();
	int size = N*N*sizeof(float);
	float* u_d;
	cudaMalloc((void**)&u_d, size);
	
	float* u_prev_d;
	cudaMalloc((void**)&u_prev_d, size);
	double stop = get_time();

	double elapsed1 = stop - start;
	std::cout<<"time to allocate in GPU= "<<elapsed1<<std::endl;

	// Copy to GPU
	start = get_time();
	cudaMemcpy( u_d, u, size, cudaMemcpyHostToDevice );
	// cudaMemcpy( u_prev_d, u_prev, size, cudaMemcpyHostToDevice );
	stop = get_time();

	double elapsed2 = stop - start;
	std::cout<<"time to copy into GPU= "<<elapsed2<<std::endl;

	// Loop 
	dim3 dimGrid(ceil(N/BLOCKSIZE),ceil(N/BLOCKSIZE),1); // number of blocks?
	dim3 dimBlock(BLOCKSIZE,BLOCKSIZE,1); // threads per block?

	start = get_time();
	for (int t=0; t<steps; t++)
	{	copy_array <<<dimGrid, dimBlock>>> (u_d, u_prev_d, N, BLOCKSIZE);

		update <<<dimGrid, dimBlock>>> (u_d, u_prev_d, N, h, dt, alpha, BLOCKSIZE);
	}
	stop = get_time();

	double elapsed3 = stop - start;
	std::cout<<"time to run kernels in GPU= "<<elapsed3<<std::endl;

	// Copy result back to host
	start = get_time();
	cudaMemcpy( u, u_d, size, cudaMemcpyDeviceToHost );
	stop = get_time();

	double elapsed4 = stop - start;
	std::cout<<"time to copy from GPU= "<<elapsed4<<std::endl;

	double total_time = elapsed1 + elapsed2 + elapsed3 + elapsed4;
	std::cout<<"total time = "<<total_time<<std::endl;
	std::cout<<"steps = "<<steps<<std::endl;

	// copy_array <<<dimGrid, dimBlock>>> (u_d, u_prev_d, N, BLOCKSIZE);
	// update <<<dimGrid, dimBlock>>> (u_d, u_prev_d, N, h, dt, alpha, BLOCKSIZE);
	// cudaMemcpy( u_h, u_prev_d, size, cudaMemcpyDeviceToHost );

	// for (int j=0; j<N; j++)
	// 	{	for (int i=0; i<N; i++)
	// 			{	I = N*j + i;
	// 				printf("%f \t %f \t %f \n",x[I], y[I], u_h[I]);
	// 			}
	// }


	std::ofstream temperature("temperature_global.txt");
	for (int j=0; j<N; j++)
	{	for (int i=0; i<N; i++)
		{	I = N*j + i;
			temperature<<x[I]<<"\t"<<y[I]<<"\t"<<u[I]<<std::endl;
		}
		temperature<<"\n";
	}

	temperature.close();

	// Free device
	cudaFree(u_d);
	cudaFree(u_prev_d);
}
```
